**常微分方程的发展历程：**
* 微积分与微分方程的诞生
* 牛顿与莱布尼茨发现微分与积分的互逆关系
* 中国在极限概念和求积的无限小方法做出贡献
* 使用微分方程研究运动与演化的物理问题（海王星的发现）
* 绝大多数常微分方程不可解，进而发展解析理论与解的性质
* 庞加莱建立常微分方程定性理论，李雅普诺夫建立稳定性理论（非线性力学）
* 伯克霍夫从定性理论抽象出动力系统的概念能，得到拓扑动力系统和遍历理论
* 定性理论和稳定性理论结合实际问题产生了最优控制理论、泛函微分方程、随机微分方程等
* 至今仍是应用数学与纯数学的活跃分支（李群与李代数的一些想法源自微分方程）

### 1.基本概念
* **微分方程**：包含未知函数及其导数或微分的关系式（等式）
* **常微分方程**：未知函数只含有一个自变量的微分方程
* **偏微分方程**：未知函数至少含有两个自变量的微分方程  

>常微分方程中的"常"指的是未知函数是一元函数  
该函数的变量通常取时间 $t$，即方程可以描述运动与演化

#### **1.1常微分方程的分类**
常微分方程$F\left(x, y, y^{\prime}, \cdots, y^{(n)}\right)=0$中，导数实际出现的最高阶数$n$叫作常微分方程的**阶**

* **n阶显式**常微分方程：方程中的$x^{(n)}$能显式地由变量$t$、未知函数$x$和它低于$n$阶的导数表示  
  $$\frac{d^{n} x}{d t^{n}}=f\left(t, x(t), x^{\prime}(t), x^{\prime \prime}(t), \cdots, x^{(n-1)}(t)\right) $$

* **n阶隐式**常微分方程：不是显式的$n$阶常微分方程
  $$\left(x^{\prime}(t)\right)^{4}+t x(t) x^{\prime}(t)+x(t)^{2}+t^{2}=0$$
- [ ] 易错点：该式为一阶隐式常微分方程

>显式相对隐式有更好的性质，比如显式的常微分方程经常存在唯一解，隐式则不然。显式与隐式之间有本质区别。
不是所有常微分方程都有显式形式，但显式方程是本课程的主要研究对象和操作对象。
 
* **n阶线性**常微分方程：方程中的函数$F$关于位置啊还能输及其各阶导数都是一次多项式
  $$x^{(n)}+a_{n-1}(t) x^{(n-1)}+\cdots+a_{1}(t) x^{\prime}+a_{0}(t) x=b_{0}(t)$$
更一般地，亦可记作：
$$a_n(t)x^{(n)}+a_{n-1}(t) x^{(n-1)}+\cdots+a_{1}(t) x^{\prime}+a_{0}(t) x=b_{0}(t),\forall t\in I,a_n(t)\not=0$$
- [ ] 易错点：$\sum\limits_{i=0}^{n}a_i(t)x^{(i)}=b_0(t)$ 在给定$t$的区间上，恒有$a_0(t)\not=0$
