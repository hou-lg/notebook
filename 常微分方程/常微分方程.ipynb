**常微分方程的发展历程：**
* 微积分与微分方程的诞生
* 牛顿与莱布尼茨发现微分与积分的互逆关系
* 中国在极限概念和求积的无限小方法做出贡献
* 使用微分方程研究运动与演化的物理问题（海王星的发现）
* 绝大多数常微分方程不可解，进而发展解析理论与解的性质
* 庞加莱建立常微分方程定性理论，李雅普诺夫建立稳定性理论（非线性力学）
* 伯克霍夫从定性理论抽象出动力系统的概念能，得到拓扑动力系统和遍历理论
* 定性理论和稳定性理论结合实际问题产生了最优控制理论、泛函微分方程、随机微分方程等
* 至今仍是应用数学与纯数学的活跃分支（李群与李代数的一些想法源自微分方程）

### 1.基本概念
* **微分方程**：包含未知函数及其导数或微分的关系式（等式）
* **常微分方程**：未知函数只含有一个自变量的微分方程
* **偏微分方程**：未知函数至少含有两个自变量的微分方程  

>常微分方程中的"常"指的是未知函数是一元函数；  
该函数的变量通常取时间 $t$，即方程可以描述运动与演化

#### **1.1常微分方程的分类**
常微分方程$F\left(x, y, y^{\prime}, \cdots, y^{(n)}\right)=0$中，导数实际出现的最高阶数$n$叫作常微分方程的**阶**

* **n阶显式**常微分方程：方程中的$x^{(n)}$能显式地由变量$t$、未知函数$x$和它低于$n$阶的导数表示  
  $$\frac{d^{n} x}{d t^{n}}=f\left(t, x(t), x^{\prime}(t), x^{\prime \prime}(t), \cdots, x^{(n-1)}(t)\right) $$

* **n阶隐式**常微分方程：不是显式的$n$阶常微分方程
  $$\left(x^{\prime}(t)\right)^{4}+t x(t) x^{\prime}(t)+x(t)^{2}+t^{2}=0$$
- [ ] 易错点：该式为一阶隐式常微分方程

>显式相较隐式有更好的性质，比如显式的常微分方程经常存在唯一解，隐式则不然。显式与隐式之间有本质区别。
不是所有常微分方程都有显式形式，但显式方程是本课程的主要研究对象和操作对象。
 
* **n阶线性**常微分方程：方程中的函数$F$关于未知函数及其各阶导数都是一次多项式
  $$x^{(n)}+a_{n-1}(t) x^{(n-1)}+\cdots+a_{1}(t) x^{\prime}+a_{0}(t) x=b_{0}(t)$$
更一般地，亦可记作：
$$a_n(t)x^{(n)}+a_{n-1}(t) x^{(n-1)}+\cdots+a_{1}(t) x^{\prime}+a_{0}(t) x=b_{0}(t),\forall t\in I,a_n(t)\not=0$$
- [ ] 易错点：$\sum\limits_{i=0}^{n}a_i(t)x^{(i)}=b_0(t)$ 在给定$t$的区间上，恒有$a_0(t)\not=0$

* **n阶非线性**常微分方程：不是线性的$n$阶常微分方程
- [ ] 易错点：判断是否为线性方程，将各函数的变量视为常数处理
- [ ] 例1：$x''+3e^{t}x'+lnt·sinx=0$ 为一阶非线性方程，因为$sinx$非一次多项式
- [ ] 例2：$x''+xx'+t=0$ 为一阶非线性方程，因为$xx'$形成了非一次项
- [ ] 例3：$x \frac{\partial f}{\partial x}+y \frac{\partial f}{\partial y}+z \frac{\partial f}{\partial z}+f=0$ 为二阶线性方程，因为$x,y,z$被视为常量处理
  
>线性系统相较非线性系统有更好的性质

本课程我们将常微分方程简称为微分方程或方程，关注其**实函数解**

#### **1.2微分方程组或微分系统**
* **微分方程组或微分系统**：多个未知函数关于一个共同的自变量及其导数满足的几个关系式组成的方程组

**【命题】**$n$阶显式常微分方程都能转化成n个未知函数的**一阶微分方程**组成的微分系统

**【证明】** 令$x_{1}=x, x_{2}=x^{\prime}, \cdots, x_{n}=x^{(n-1)}$，则：
$$
\left\{\begin{array}{ll}
x_{1}^{\prime} & =x_{2}, \\
x_{2}^{\prime} & =x_{3}, \\
\vdots \\
x_{n-1}^{\prime} & =x_{n}, \\
x_{n}^{\prime} & =f\left(t, x_{1}, x_{2}, \cdots, x_{n}\right)
\end{array}\right.
$$
>显式常微分方程亦可根据$f$的表达式构造方程组；
但此法对隐式无效，无法实现最后一个式子

* **微分系统的维数**：若微分系统都是由**一阶微分方程**组成的，且这个微分系统中<u>未知函数的个数是n</u>，则称它为**n维微分系统**

* **n维线性微分系统**：若$\mathbf{x}(t)$表示$n$维实向量函数，$\mathbf{x}(t)=\left(x_{i}(t)\right)_{n \times 1} \in \mathbb{R}^{n}$
    则$n$维线性微分线性系统可表示为：
  $$\frac{d \mathbf{x}(t)}{d t}=A(t) \mathbf{x}(t)+\mathbf{b}(t), A(t)=\left(a_{i j}(t)\right)_{n \times n}, \mathbf{b}(t)=\left(b_{i}(t)\right)_{n \times 1}$$

>$n$维 $\implies$系统中的方程都是一阶，最高只有一阶导数
线性$\implies$n个未知函数及其导函数均为一次多项式，导函数可分离到等式一端

**【命题】**$n$阶线性微分方程 $\implies n$维线性微分系统

**【证明】** $n$阶线性微分方程有显式形式 $x^{(n)}=f\left(t, x(t), x^{\prime}(t), x^{\prime \prime}(t), \cdots, x^{(n-1)}(t)\right) $  
设 $x_{1}=x, x_{2}=x^{\prime}, \cdots, x_{n}=x^{(n-1)}$，则：
$$
\left\{\begin{array}{ll}
x_{1}^{\prime} & =x_{2}, \\
x_{2}^{\prime} & =x_{3}, \\
\vdots \\
x_{n-1}^{\prime} & =x_{n}, \\
x_{n}^{\prime} & =f\left(t, x_{1}, x_{2}, \cdots, x_{n}\right)
\end{array}\right.
$$
其中$x_n'=f\left(t, x_{1}, x_{2}, \cdots, x_{n}\right)$亦为一阶线性方程，故所得为n阶线性微分系统

**【思考】** 是否有：$n$维线性微分系统$\implies n$阶线性微分方程？限制条件？
**【命题】** （想法）$n$维线性微分系统可化为$n$阶线性微分方程$\iff r\left(A(t)\right)=1$
**【证明】** （1）若$r\left(A(t)\right)=1$，则经过初等变换（左乘矩阵）可得
$$
\left\{\begin{array}{ll}
c_1x_{1}^{\prime} &=a_1x_n+b_1(t) \\
c_2x_{2}^{\prime} &=a_2x_n+b_2(t) \\
\vdots \\
c_{n-1}x_{n-1}^{\prime} &=a_{n-1}x_{n}+b_{n-1}(t) \\
c_{n}x_{n}^{\prime} & =a_{n}x_{n}+b_{n}(t)
\end{array}\right.
$$
？？？未解决

#### **1.3微分方程的解**
* **微分方程的一个解**
设函数 $ x=\phi(t) $ 在区间 $ J $ 上**连续**,  $J \subset I \subset \mathbb{R}$ , 且它的一阶 至  $n-1$  阶导数**连续可微**. 如果 $ x=\phi(t)$  在区间 $ J $ 上满 足微分方程$F\left(t, x, x^{\prime}, \cdots, x^{(n)}\right)=0$, 即: 把 $ x=\phi(t)$  及其一阶至 $ n$  阶导数, 带入 $ F\left(t, x, x^{\prime}, \cdots, x^{(n)}\right) $ 得到
$$
F\left(t, \phi(t), \phi^{\prime}(t), \cdots, \phi^{(n)}(t)\right) \equiv 0, \forall t \in J,
$$
则称  $x=\phi(t)$  是该微分方程在区间  $J$  上的一个解.
- [ ] 易错点：微分方程的解是“未知函数”，包含定义域，值域，规则 三部分
- [ ] 易错点：给定微分方程，即给定了$n+2$元函数的定义域 $I \times{J_1} \times{J_2} \cdots \times{J_n}$
- [ ] 易错点：函数表达式限制自变量$t$的区间$I$，是未知函数$x$的最大定义域

* **微分方程的通解**
设  $c_{1}, c_{2}, \cdots, c_{n}$  是 $n$ 个任意常数, 如果函数 $ x=\phi\left(t, c_{1}, c_{2}, \cdots, c_{n}\right) $ 是区间 $J$ 上微分方程$ F\left(t, x, x^{\prime}, \cdots, x^{(n)}\right) $的解，并且这 $ n $ 个任意常数 $ c_{1}, c_{2}, \cdots, c_{n}$  是**独立的**,即: 向量函数  $\left(\phi\left(t, c_{1}, c_{2}, \cdots, c_{n}\right), \phi^{\prime}\left(t, c_{1}, c_{2}, \cdots, c_{n}\right), \cdots\right. , \left.\phi^{(n-1)}\left(t, c_{1}, c_{2}, \cdots, c_{n}\right)\right) $ 关于向量$  \left(c_{1}, c_{2}, \cdots, c_{n}\right)  $的 **Jacobi 行列式**在区间 $ J $ 上不等于零,
$$
\frac{D\left[\phi, \phi^{\prime}, \cdots, \phi^{(n-1)}\right]}{D\left[c_{1}, c_{2}, \cdots, c_{n}\right]} \triangleq\left|\left(\begin{array}{ccc}
\frac{\partial \phi}{\partial c_{1}} & \cdots & \frac{\partial \phi}{\partial c_{n}} \\
\vdots & \vdots & \vdots \\
\frac{\partial \phi^{(n-1)}}{\partial c_{1}} & \cdots & \frac{\partial \phi^{(n-1)}}{\partial c_{n}}
\end{array}\right)\right| \neq 0, \forall t \in J
$$
则称 $ x=\phi\left(t, c_{1}, c_{2}, \cdots, c_{n}\right)  $是该微分方程在区间 $ J$  上的通解.
